```python
### Exercise 1

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Dataset
house_sizes = np.array([50, 70, 80, 100, 120]).reshape(-1, 1)  # Reshape for sklearn
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Initialize and fit the model
model = LinearRegression()
model.fit(house_sizes, house_prices)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Predict the price of a house that is 90 square meters
house_size_90 = np.array([[90]])
predicted_price_90 = model.predict(house_size_90)

# Print the results
print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
print(f"Predicted price of a 90 square meter house: {predicted_price_90[0]}")
```

```python
# Plotting the data and the regression line
plt.scatter(house_sizes, house_prices, color='blue', label='Actual Prices')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Regression Line')
plt.scatter(house_size_90, predicted_price_90, color='green', marker='x', s=100, label='Predicted Price for 90 sqm')
plt.xlabel('House Size (sqm)')
plt.ylabel('House Price (currency units)')
plt.title('House Prices vs. Size with Linear Regression')
plt.legend()
plt.show()

# Interpretation of the slope
slope_interpretation = f"""
The slope of the regression line is {slope}. This means that for every additional square meter increase in the size of the house, the price of the house increases by {slope} currency units.
In other words, if the size of the house increases by 1 square meter, the house price is expected to increase by {slope} currency units, holding all other factors constant.
"""

print(slope_interpretation)
```

```python
### Exercise 2
from scipy import stats

# Dataset
fertilizer_1 = np.array([5, 6, 7, 6, 5])
fertilizer_2 = np.array([7, 8, 7, 9, 8])
fertilizer_3 = np.array([4, 5, 4, 3, 4])

# Perform ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Print the results
print(f"F-value: {f_value}")
print(f"P-value: {p_value}")

# Interpretation based on P-value
if p_value < 0.05:
    print("The P-value is less than 0.05, indicating that there are significant differences between the effects of the fertilizers on plant growth.")
else:
    print("The P-value is greater than 0.05, indicating that there are no significant differences between the effects of the fertilizers on plant growth.")

# Explanation of P-value greater than 0.05
if p_value > 0.05:
    explanation = """
    If the P-value were greater than 0.05, we would fail to reject the null hypothesis. This means that there is no statistically significant evidence to suggest that the different fertilizers have different effects on plant growth.
    In other words, any observed differences in plant growth could be due to random variation rather than the effect of the fertilizers.
    """
    print(explanation)
```

```python
### Exercise 3

# Parameters for the binomial distribution
n = 10  # Number of trials (coin flips)
p = 0.5  # Probability of success (getting a head)

# Calculate the probability of getting exactly 5 heads in 10 coin flips
k = 5  # Number of successes (heads)
probability = stats.binom.pmf(k, n, p)

print(f"The probability of getting exactly {k} heads in {n} coin flips is {probability:.4f}")
```

```python
### Exercise 4
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Example dataset
data1 = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Extracting the variables
age = data1['age']
income = data1['income']

# Calculate Pearson correlation
pearson_corr, pearson_p_value = pearsonr(age, income)

# Calculate Spearman correlation
spearman_corr, spearman_p_value = spearmanr(age, income)

# Print the results
print(f"Pearson correlation coefficient: {pearson_corr:.4f}, P-value: {pearson_p_value:.4f}")
print(f"Spearman correlation coefficient: {spearman_corr:.4f}, P-value: {spearman_p_value:.4f}")
```

```python
### Exercise 5

# Example dataset
data2 = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Extracting the variables
age = data2['age'].values.reshape(-1, 1)
income = data2['income'].values

# Create a scatter plot
plt.scatter(data2['age'], data2['income'], color='blue', label='Data points')

# Fit the linear regression model
model = LinearRegression()
model.fit(age, income)

# Predict the income values based on the age values
income_pred = model.predict(age)

# Plot the regression line
plt.plot(age, income_pred, color='red', label='Regression line')

# Add titles and labels
plt.title('Scatter Plot of Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()

# Show plot
plt.show()
```
